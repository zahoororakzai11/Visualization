```python
import pandas as pd
import numpy as np
# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
# print the graphs in the notebook
%matplotlib inline
# set seaborn style to white
sns.set_style("white")
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
tips = pd.read_csv(url)

tips.head()
```

```python
del tips['Unnamed: 0']
tips.head()
```

```python
tot_bill = sns.distplot(tips['total_bill'])

# set the label
tot_bill.set(xlabel='Value',ylabel='Frequence',title='Total Bill')

# take the upper and right borders
sns.despine()
```

```python
#scatter plot to show relationship between total_bill and tip
sns.jointplot(x = 'total_bill',y='tip',data = tips)
```

```python
# relationship b/w varialbe
sns.pairplot(tips)
```

```python
# realtionship b/w numberical and categorial variable
sns.stripplot(x = 'day',y = 'total_bill',data = tips,jitter = True)
```

```python
sns.stripplot(x = 'tip',y = 'day',hue = 'sex',data = tips,jitter = True)
```

```python
# boxplot
sns.boxplot(x = 'day',y = 'total_bill',hue = 'time',data = tips)
```

```python
# two histogram
sns.set(style='ticks')

#created FacetGrid
g= sns.FacetGrid(tips,col='time')
g.map(plt.hist,'tip')
```

```python
a = sns.FacetGrid(tips,col='sex',hue= 'smoker')
a.map(plt.scatter,'total_bill','tip',alpha=.7)
a.add_legend();
```

## Titanic_Disaster Dataset

```python

data = pd.read_csv('/content/sample_data/tested.csv')
data.head()
```

```python
data.set_index('PassengerId').head()
```

```python
# pie to presenting male/female proportion
males = (data['Sex']=='male').sum()
females = (data['Sex'] =='female').sum()

# put them into a list called proportions
proportions = [males, females]

plt.pie(proportions,labels = ['Males','Females'],shadow = True,colors=['blue','red'],
        startangle = 0, explode = (0.15,0),autopct = '%1.1f%%')
plt.axis('equal')
plt.title('Sex Proportion')
plt.tight_layout()
plt.show()
```

```python
lm = sns.lmplot(x = 'Age',y = 'Fare',data= data,hue = 'Sex',fit_reg=False)
lm.set(title = 'Fare x Age')
axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```
