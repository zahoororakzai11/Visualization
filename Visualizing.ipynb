```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

#set this so the
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
chipo.head(4)
```

### Histogram

```python
#Creat a histogram on top 5 items bought
x = chipo.item_name
# use a counter class from collections to create a dictionary
letter_counts = Counter(x)

#convert the dictionary to DataFrame
df = pd.DataFrame.from_dict(letter_counts,orient ='index')

#sort the value from top to the least value and slice
df = df[0].sort_values(ascending= True)[45:50]

#create plot
df.plot(kind= 'bar')

#Set the title and lables
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title("Most ordered Chippotle \s Items")

#show the plot
plt.show()
```

```python
# Scatterplot with number of items rdered per order price

#creat a list of price
chipo.item_price = [float(value[1:-1].strip()) for value in chipo.item_price]

#group the order and sum
orders = chipo.groupby('order_id').sum()

#creat the scatterplot
#plt.scatterplot(order.quantity,orders.item_price,s = 5)
plt.scatter(x = orders.item_price,y= orders.quantity,s = 50,c='green')

#Set the title and lables
plt.xlabel('Order Price')
plt.ylabel('Items Ordered')
plt.title("Number of item orderd per order price")
plt.ylim(0)
```
