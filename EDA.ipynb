```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/GP Orders - 4.csv',encoding='utf-8')
df.head()
```

```python
df1 = pd.read_csv('/content/sample_data/GP Orders - 5.csv',encoding='utf-8')
df1.head()
```

```python
#join df with df1
tables = df.append(df1,ignore_index=True)
tables
```

```python
# check Null values
tables.isna().sum()
```

```python
tables.duplicated().value_counts()
```

```python
#Change Name of Column
tables.rename(columns={'City (Billing)':'City'},inplace=True)
tables
```

```python
Table_csv = df.to_csv('table.csv')
Table = pd.read_csv('table.csv')
Table
```

```python
# Top 10 Books Most Selling Books by Publisher
top_10_book = Table.groupby('Book Name')['Order Number'].agg(["count"]).reset_index().sort_values(by='count',ascending=False).head(10)
top_10_book
```

```python
# Visualization
plt.figure(figsize=(10,5))
graph = sns.barplot(x='Book Name',y='count',data=top_10_book)
graph.set_title("Top 10 Selling Books")
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 least Selling Books
Topleastbook=Table.groupby('Book Name')['Order Number'].agg(['count']).reset_index().sort_values(by='count',ascending=False).tail(10)
print(Topleastbook)
```

```python
plt.figure(figsize=(10,5))
graph = sns.barplot(x='Book Name',y= 'count',data= Topleastbook)
graph.set_title('Top 10 Least Book Selling')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top City Buy Most Books
Top_10_city=Table.groupby("City")["Order Number"].agg(["count"]).reset_index().sort_values(by="count",ascending=False).head(10)
Top_10_city
```

```python
plt.figure(figsize=(12,5))
graph=sns.barplot(x="City",y="count",data=Top_10_city)
graph.set_title("Top 10 Cities In whcih Book Sold Out")
plt.xticks(rotation=90)
plt.show()
```

```python
# order status of book

order = Table.groupby('Order Status')['Order Number'].agg(['count']).reset_index()
order
```

```python
graph = sns.barplot(x= 'Order Status',y='count',data=order)
graph.set_title("Order Status")
plt.show()
```

```python
# Payment Method
Changepay = {'Cash on Delivery':'Cash on delivery'}
Table['Payment Method']=Table['Payment Method'].replace(Changepay)
paymethod= Table.groupby('Payment Method')['Order Number'].agg(['count']).reset_index()
paymethod
```

```python
plt.figure(figsize=(10,5))
graph= sns.barplot(x="Payment Method",y="count",data=paymethod)
graph.set_title("Payment Method")
plt.show()
```

```python
# Analyze Date of order Status of Books
Completedorder = Table[Table['Order Status'] == 'Completed']
Completedorder = Completedorder.groupby('City')['Order Number'].agg(['count']).reset_index().sort_values(by='count',ascending=False).head(10)
Completedorder
```

```python
# Visualization them
plt.figure(figsize=(10,5))
graph = sns.barplot(x = 'City',y = 'count',data = Completedorder)
graph.set_title("Book Order Completed")
plt.show()
```
