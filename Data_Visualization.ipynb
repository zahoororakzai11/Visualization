```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Warning

import warnings
warnings.filterwarnings('ignore')
```

```python
tips_data = sns.load_dataset('tips')
tips_data.head(5)
```

# 1.1 Line Plot
To show relationship between dependent and independent variable in dataset

```python
import numpy as np
x = np.linspace(1,20,100)
y = np.exp(x)

#line plot
plt.plot(x,y)


#display the plot
plt.show()
```

```python
# data
X = np.linspace(1,20,100)
Y = np.exp(X)

# the argument 'r*' plots each point as a red '*'
plt.plot(X,Y,'r*')
plt.show()
```

```python
# plot multiple line
X = np.linspace(1,20,100)
Y1 = X
Y2 = np.square(X)
Y3 = np.sqrt(X)

#line Plot
plt.plot(X,Y1,'r',X,Y2,'b',X,Y3,'g')
#add title to the plot
plt.title('Line Plot')

#display the plot
plt.show()
```

# 1.2 Scatter Plot
plotted on horizontal and vertical axes,to show correlation between two varialble,One can detect extreme data points using scatter plot

```python
X = tips_data['total_bill']
Y = tips_data['tip']

#plot the scatter plot
plt.scatter(X,Y)

#add the axes lable to plot
plt.xlabel('total_bill')
plt.ylabel('tips')

#display the plot
plt.show()
```

```python
# Plot scatter plot for 'total_bill'
x = tips_data['total_bill']
y = tips_data['tip']
#plot the scatter plot
# s for shape,c is for color, alpha is for opaacity
plt.scatter(x,y,s = np.array(y)**2,c = 'green',alpha=0.8)
# add title
plt.title('Scatter Plot')

# add the axes labels to the plot
plt.xlabel('total_bill')
plt.ylabel('tip')

# display the plot
plt.show()
```

# 1.6 Box Plot

```python
plt.boxplot(tips_data['total_bill'])

#add label for five number summary
plt.text(x = 1.1, y= tips_data['total_bill'].min(),s='min')
plt.text(x= 1.1,y = tips_data.total_bill.quantile(0.25),s = 'Q1')
plt.text(x = 1.1,y= tips_data['total_bill'].median(),s='median(Q2)')
plt.text(x = 1.1,y = tips_data.total_bill.quantile(0.75),s = 'Q3')
plt.text(x = 1.1, y = tips_data['total_bill'].max(),s= 'max')

#add an arrow (annonate ) to show outliers
plt.annotate('Outliers',xy = (0.97,45),xytext=(0.7,44), arrowprops = dict(facecolor = 'black',arrowstyle = 'simple'))
# add an arrow (annonate) to show the outliers
#plt.annotate('Outliers', xy = (0.97,45),xytext=(0.7, 44), arrowprops = dict(facecolor = 'black', arrowstyle = 'simple'))


# add the graph title and axes labels
plt.title('Boxplot of Total Bill Amount')
plt.ylabel('Total bill')

# display the plot
plt.show()
```

# 2.1 Strip plot

```python
#check the head() fo tips dataset
tips_data.head()
```

```python
# strip plot
sns.stripplot(y='tip',x = 'time',data = tips_data)

#display the plot
plt.show()
```

```python
# to display the spread of the data,jitter points
sns.stripplot(x = 'time',y = 'tip',data = tips_data,jitter = True)

#display the plot
plt.show()
```

# 2.2 Swarm Plot
To Avoid the overlapping of dataponts

```python
# Swarm plot
sns.swarmplot(x = 'time',y= 'tip',data = tips_data)

#display the plot
plt.show()
```

```python
# can add hue to show categorial clearly
sns.swarmplot(x = 'time',y = 'tip', data= tips_data,hue = 'day')

#display the plot
plt.show()
```

# 2.3 Violine Plot
Show relationship b/w numberical variable across different categorical variable.


```python
# check head of the tips dataset
tips_data.head()
```

```python
# violine plot of numberical variable 'total_bill' and categorial variable day
sns.violinplot(x = 'day',y = 'total_bill',data = tips_data)

# display plot
plt.show()
```

```python
# add more variable
sns.violinplot(y = 'total_bill' , x = 'day', data = tips_data,hue = 'sex',split = True)

#display
plt.show()
```

```python
# Distribution Plot
sns.displot(tips_data['total_bill'], kind  = 'kde')

plt.show()
```

```python
# # iterate the distplot() function over the time

# # list of time
time = ['Lunch', 'Dinner']

# # iterate through time
for i in time:
    subset = tips_data[tips_data['time'] == i]

# # Draw the density plot`
# # 'hist = False' will not plot a histogram
# # 'kde = True' plots density curve
    sns.distplot(subset['total_bill'], hist = False, kde = True,
                  kde_kws = {'shade':True},
                  label = i)
```

```python
# Count plot

sns.countplot(data = tips_data, x = 'day',hue = 'time')

plt.show()
```
